```python
##Calculate the percentage of the missing values of all the columns
cols=curr_app.columns
ind=len(curr_app.index)
colm=[]
percnt=[]
Null=[]
tot=[]
for col in cols:
    colm.append(col)
    null=curr_app[col].isna().sum()
    Null.append(null)
    tot.append(ind)
    per=null*100/ind
    percnt.append(per)
colm=pd.Series(colm)
percnt=pd.Series(percnt)
Null=pd.Series(Null)
total=pd.Series(tot)
missing_val=pd.concat([colm, Null,total,percnt], axis=1, keys=['Column', 'Null Values','Total Values','Percent of NULL Values']).sort_values(by='Percent of NULL Values',ascending=False)#.set_index(keys='Column',inplace=True)
missing_val.set_index('Column')
missing_val
```

```python
##Drop the columns where % of missing values > 50
cols=curr_app.columns
ind=len(curr_app.index)
NR_cols=[]
for col in cols:
    null=curr_app[col].isna().sum()
    per=null*100/ind
    if per > 50:
        NR_cols.append(col)
#for col in NR_cols:
curr_app=curr_app.drop(NR_cols, axis = 1)
#curr_app
```

## Perform the Univariate Analysis for TARGET value 1 data set
#### There are 2 types of viables, Catogical (Dimensions) and contineous (Measures) 
#### Categorical vriables can be further devided as 2 types, ordered and un-ordered categorical variables
#### So we will perform the Univariate analysis on all these kind of vriables

### Below is the list of un-ordered Categorial varables having Boolean values, lets see the univariate analysis of these

```python
def uni_var_payment_diff(x):
    #from plotly.subplots import make_subplots
    
    temp = payment_diff[x].value_counts()
    df1 = pd.DataFrame({x: temp.index,'Frequency': temp.values})
    df1.set_index(x,inplace=True)
    ax = df1.plot(kind='bar',color='red')
    percent = ["{}%".format(int(100.*row.Frequency/sum(df1['Frequency']))) for name,row in df1.iterrows()]
    print(percent)
   
    for i,child in enumerate(ax.get_children()[:df1.index.size]): 
        ax.text(i,child.get_bbox().y1+200,percent[i], horizontalalignment ='center')
        ax.get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x))))
        ax.patch.set_facecolor('#FFFFFF')
        ax.spines['bottom'].set_color('#CCCCCC')
        ax.spines['bottom'].set_linewidth(1)
        ax.spines['left'].set_color('#CCCCCC')
        ax.spines['left'].set_linewidth(1)
    plt.title("Customers with Payment Diffculty")
    plt.show()

def uni_var_no_payment_diff(x):
    #from plotly.subplots import make_subplots
    
    temp = payment_no_diff[x].value_counts()
    df1 = pd.DataFrame({x: temp.index,'Frequency': temp.values})
    df1.set_index(x,inplace=True)
    ax = df1.plot(kind='bar')
    percent = ["{}%".format(int(100.*row.Frequency/sum(df1['Frequency']))) for name,row in df1.iterrows()]
    #print(percent)
   
    for i,child in enumerate(ax.get_children()[:df1.index.size]): 
        ax.text(i,child.get_bbox().y1+200,percent[i], horizontalalignment ='center')
        ax.get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x))))
        ax.patch.set_facecolor('#FFFFFF')
        ax.spines['bottom'].set_color('#CCCCCC')
        ax.spines['bottom'].set_linewidth(1)
        ax.spines['left'].set_color('#CCCCCC')
        ax.spines['left'].set_linewidth(1)
    plt.title("Customers with No Payment Diffculty")
    plt.show()

def uni_var_analysis(x):
    uni_var_payment_diff(x)
    uni_var_no_payment_diff(x)
```

```python
def uni_var_final_payment_diff(x):
    #from plotly.subplots import make_subplots
    
    temp = final_payment_diff[x].value_counts()
    df1 = pd.DataFrame({x: temp.index,'Frequency': temp.values})
    df1.set_index(x,inplace=True)
    ax = df1.plot(kind='bar',color='red')
    percent = ["{}%".format(int(100.*row.Frequency/sum(df1['Frequency']))) for name,row in df1.iterrows()]
    print(percent)
   
    for i,child in enumerate(ax.get_children()[:df1.index.size]): 
        ax.text(i,child.get_bbox().y1+200,percent[i], horizontalalignment ='center')
        ax.get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x))))
        ax.patch.set_facecolor('#FFFFFF')
        ax.spines['bottom'].set_color('#CCCCCC')
        ax.spines['bottom'].set_linewidth(1)
        ax.spines['left'].set_color('#CCCCCC')
        ax.spines['left'].set_linewidth(1)
    plt.title("Customers with Payment Diffculty")
    plt.show()

def uni_var_no_final_payment_diff(x):
    #from plotly.subplots import make_subplots
    
    temp = final_payment_no_diff[x].value_counts()
    df1 = pd.DataFrame({x: temp.index,'Frequency': temp.values})
    df1.set_index(x,inplace=True)
    ax = df1.plot(kind='bar')
    percent = ["{}%".format(int(100.*row.Frequency/sum(df1['Frequency']))) for name,row in df1.iterrows()]
    #print(percent)
   
    for i,child in enumerate(ax.get_children()[:df1.index.size]): 
        ax.text(i,child.get_bbox().y1+200,percent[i], horizontalalignment ='center')
        ax.get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x))))
        ax.patch.set_facecolor('#FFFFFF')
        ax.spines['bottom'].set_color('#CCCCCC')
        ax.spines['bottom'].set_linewidth(1)
        ax.spines['left'].set_color('#CCCCCC')
        ax.spines['left'].set_linewidth(1)
    plt.title("Customers with No Payment Diffculty")
    plt.show()

def final_uni_var_analysis(x):
    uni_var_final_payment_diff(x)
    uni_var_no_final_payment_diff(x)
```
